# Movie Profit Analysis

The dataset we will use is from the Data and Story Library, using information from IMDB.com, Boxofficemojo.com, and Rottentomatoes.com.
The dataset is found at this link: https://dasl.datadescription.com/datafile/movie-profits/

The dataset looks at a sample of movies from 2008 to 2012 and compares their run time, critic score on Rotten Tomatoes, and their budget with their US Gross Revenue.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
# https://dasl.datadescription.com/download/data/3349
df=pd.read_csv("https://dasl.datadescription.com/download/data/3349", delimiter="\t")
df.columns=['Year','Movie','US Gross Revenue (mil $)','Budget (mil $)','Runtime (min)','Critic Score (%)']
```

```python
dfdata1=df[['US Gross Revenue (mil $)','Budget (mil $)']]
```

```python
dfdata1.corr()
```

```python
dfdata1.columns=['revenue','budget']
```

```python
plt.scatter(dfdata1.revenue, dfdata1.budget)
plt.xlabel('US Gross Revenue (mil $)')
plt.ylabel('Budget (mil $)')
plt.show()
```

```python
dfdata1cut = dfdata1[dfdata1.revenue <= 250]
```

```python
plt.scatter(dfdata1cut.revenue, dfdata1cut.budget)
plt.xlabel('US Gross Revenue (mil $)')
plt.ylabel('Budget (mil $)')
plt.show()
```

```python
plt.hist(dfdata1cut.revenue)
plt.title('US Gross Revenue (mil $)')
plt.show()
plt.hist(dfdata1cut.budget)
plt.title('Budget (mil $)')
plt.show()
```

```python
dfdata2=df[['US Gross Revenue (mil $)','Critic Score (%)']]
```

```python
dfdata2.corr()
```

```python
dfdata2.columns=['revenue','score']
```

```python
plt.scatter(dfdata2.revenue, dfdata2.score)
plt.xlabel('US Gross Revenue (mil $)')
plt.ylabel('Critic Score (%)')
plt.show()
```

```python
dfdata2cut = dfdata2[dfdata2.revenue <= 250]
```

```python
plt.scatter(dfdata2cut.revenue, dfdata2cut.score)
plt.xlabel('US Gross Revenue (mil $)')
plt.ylabel('Critic Score (%)')
plt.show()
```

```python
plt.hist(dfdata2cut.revenue)
plt.title('US Gross Revenue (mil $)')
plt.show()
plt.hist(dfdata2cut.score)
plt.title('Critic Score (%)')
plt.show()
```

I decided to compare revenue and budget of the list films, which have a clear correlation (.683109), which is also shown in the graphics, as well as comparing critic score and revenue, which have a much weaker correlation (.346443).
