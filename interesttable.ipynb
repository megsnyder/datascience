# Interest Table
This is a table and graph for the value of an investment after x years with an initial value of 1000 dollars compounded at 3.5\% annually using the equation $v=1000(1.035)^{x}$

```python
from numpy import radians, sin
from numpy import linspace
import matplotlib.pyplot as plt
import pandas as pd
```

```python
interesttable = [(x,(1000*(1.035)**x)) for x in linspace(0,30,31)]
```

```python
interesttable = pd.DataFrame()
interesttable ["years"] = linspace(0,30,31)
interesttable ["value_float"] = [1000*(1.035)**x for x in linspace(0,30,31)]
interesttable ["value_str"] = ['$' + str(int(1000*(1.035)**x)) for x in linspace(0,30,31)]
```

```python
interesttable.columns = ['years','value_float','value_str']
```

```python
interesttable
```

```python
plt.plot(interesttable.years, interesttable.value_float)
plt.xlabel(interesttable.years.name)
plt.ylabel(interesttable.value_float.name)
plt.title('Value of investment as a function of years at intial investment of $1000 with a rate of 3.5% compounded annually')
plt.legend()
plt.show()
```
